```python
# Import dependencies and setup

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Load .csv - 'Value of new borrower-accepted loan commitments'

value_loan_commitments_df = pd.read_csv('./Resources/value_new_borrower_accepted_loans.csv')
value_loan_commitments_df
```

```python
# Plot 'Value of new borrower-accepted loan commitments'

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plot
plt.figure(figsize=(15,5))
plt.bar(value_loan_commitments_df['Sector'], value_loan_commitments_df['March 2022 ($b)'], color="blue", 
        width=0.5, alpha=0.75)

# Title / labels
plt.title("Value of new borrower-accepted loan commitments, seasonally adjusted, March 2022",  fontdict = font_1)
plt.ylabel("$ Billion",  fontdict = font_2)

# Grid
matplotlib.pyplot.grid(visible=True, which='major', axis='y', color='blue', alpha=0.05, linestyle='-', linewidth=1)

# Aesthetics
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Show/save
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/australian-lending-indicators-march-2022/Images/new_borrower_accepted_loans.png", 
            transparent=False, dpi=300)
plt.show()
```
