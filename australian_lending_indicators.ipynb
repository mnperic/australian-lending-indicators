```python
# Import dependencies and setup

import pandas as pd
import matplotlib.pyplot as plt
import matplotlib
```

```python
# Load .csv - 'Value of new borrower-accepted loan commitments'

value_loan_commitments_df = pd.read_csv('./Resources/value_new_borrower_accepted_loans.csv')
value_loan_commitments_df
```

```python
# Plot 'Value of new borrower-accepted loan commitments'

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Plot
plt.figure(figsize=(15,5))
plt.bar(value_loan_commitments_df['Sector'], value_loan_commitments_df['March 2022 ($b)'], color="blue", 
        width=0.5, alpha=0.75)

# Title / labels
plt.title("Value of new borrower-accepted loan commitments, seasonally adjusted, March 2022",  fontdict = font_1)
plt.ylabel("$ Billion",  fontdict = font_2)

# Grid
matplotlib.pyplot.grid(visible=True, which='major', axis='y', color='blue', alpha=0.05, linestyle='-', linewidth=1)

# Aesthetics
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Show/save
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/australian-lending-indicators-march-2022/Images/new_borrower_accepted_loans.png", 
            transparent=False, dpi=300)
plt.show()
```

```python
# Convert (%) columns to float

value_loan_commitments_df['Monthly percent change (%)'] = value_loan_commitments_df['Monthly percent change (%)'].astype(float)
value_loan_commitments_df['Yearly percent change (%)'] = value_loan_commitments_df['Yearly percent change (%)'].astype(float)
```

```python
# Load .csv - 'new loan commitments, housing'

housing_loans_df = pd.read_csv('./Resources/new_loan_commitments_housing_purpose.csv')
housing_loans_df
```

```python
# Convert (%) columns to float

housing_loans_df['Owner occupier (ex. refinancing)'] = housing_loans_df['Owner occupier (ex. refinancing)'].astype(float)
housing_loans_df['Investor (ex. refinancing)'] = housing_loans_df['Investor (ex. refinancing)'].astype(float)
```

```python
# Set 'Date' as index

housing_loans_index_df = housing_loans_df.set_index('Date')
```

```python
# Plot - housing loans

# Font families
font_1 = {'family':'helvetica','color':'black','size':16}
font_2 = {'family':'helvetica','color':'black','size':12}

# Define data
housing_loans_data = housing_loans_index_df[['Owner occupier (ex. refinancing)', 'Investor (ex. refinancing)']]                

# Plot
housing_loans_data.plot(figsize=(15,5), kind="bar", stacked=False, alpha=0.9, color=['red', 'blue'], rot=45
                                                                
             ).legend(ncol=1)

plt.title("New loan commitments for housing by purpose (seasonally adjusted), monthly growth, 2020–2022", 
          fontdict = font_1)
plt.xlabel("Year", fontdict = font_2)
plt.ylabel("% month-to-month movement", fontdict = font_2)

# Grid
matplotlib.pyplot.grid(visible=True, which='major', axis='both', color='black', alpha=0.05, linestyle='-', linewidth=1)

# Aesthetics
plt.gca().spines['top'].set_visible(False)
plt.gca().spines['right'].set_visible(False)
plt.tight_layout()

# Show/save
plt.savefig("/Users/minoperic/Data Analytics/Data Investigations/australian-lending-indicators-march-2022/Images/housing_loans.png", 
            transparent=False, dpi=300)
plt.show()
```
